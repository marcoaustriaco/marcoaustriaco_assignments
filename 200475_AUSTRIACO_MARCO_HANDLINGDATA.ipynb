```python
def relationship_status(from_member, to_member, social_graph):
    check1 = social_graph.get(from_member) 
    check2 = check1.get("following") #checks who from_member follows
    check3 = social_graph.get(to_member)
    check4 = check3.get("following") #checks who to_member follows
    for i in check2:
        if i == to_member:
            for j in check4:
                if j == from_member:
                    return "friends"
                else:
                    continue
            return "follower"        
        else:
            continue
    for k in check4:
        if k == from_member:
            return "followed by"
        else:
            continue
    return None


relationship_status("@eeebeee","@jobenilagan",social_graph)
```

```python
def tic_tac_toe(board):
    
    #check horizontal
    for i in range(len(board)):
        xcounter = 0
        ocounter = 0
        row = board[i]
        for j in range(len(row)):
            if row[j] == "X":
                xcounter += 1
                if xcounter == len(board):
                    return "X"
                else:
                    continue
            elif row[j] == "O":
                ocounter += 1
                if ocounter == len(board):
                    return "O"
                else:
                    continue
            else:
                continue
    
    #check vertical
    for k in zip(*board):
        xcounter = 0
        ocounter = 0
        column = k
        for l in range(len(k)):
            if column[l] == "X":
                xcounter += 1
                if xcounter == len(board):
                    return "X"
                else:
                    continue
            elif column[l] == "O":
                ocounter += 1
                if ocounter == len(board):
                    return "O"
                else:
                    continue
            else:
                continue

    
    #check diagonal left to right
    xcounter = 0
    ocounter = 0
    current = 0
    for m in range(len(board)):
        if board[m][current] == "X":
            xcounter += 1
            current += 1
            if xcounter == len(board):
                return "X"
            else:
                continue
        elif board[m][current] == "O":
            ocounter += 1
            current += 1
            if ocounter == len(board):
                return "O"
            else:
                continue
    
    #check diagonal right to left
    xcounter = 0
    ocounter = 0
    current1 = len(board)-1
    for m in range(len(board)):
        if board[m][current1] == "X":
            xcounter += 1
            current1 -= 1
            if xcounter == len(board):
                return "X"
            else:
                continue
        elif board[m][current1] == "O":
            ocounter += 1
            current1 -= 1
            if ocounter == len(board):
                return "O"
            else:
                continue
                
    return None


            
tic_tac_toe(board7)
#REMOVE DIRECTIONS BEFORE SUBMITTING!!       
    
    
            
        
        
    
```

# Problem 3:  Routing

```python
legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}
```

```python
def eta(legs,source,destination):
    starting = [i[0] for i in legs] #list all starting points
    ending = [j[1] for j in legs] #list all ending points
    collection = list(legs.values()) #all the times
    
    total_time = 0
        
    for k in range(len(legs)):
        if k >= starting.index(source):
            total_time += collection[starting.index(starting[k])]["travel_time_mins"]
            if k == ending.index(destination):
                break
            else:
                continue
        else:
            continue
        
    if k != ending.index(destination): #LOOPING!
        for m in range(0,len(legs)):
            total_time += collection[ending.index(ending[m])]["travel_time_mins"]
            if ending[m] == destination:
                break
            continue
            
    return total_time 

            
eta(legs,"dlsu","dlsu")
```
